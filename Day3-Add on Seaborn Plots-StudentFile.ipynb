```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['patch.force_edgecolor']=True
```

```python
movies = pd.read_csv(r"/Users/dhrubajyotibiswas/codebase/datascience/ProgrammingForDataScience/MLData/Movie-Ratings.csv")

movies.columns = (['Film','Genre','CriticsRatings','AudienceRatings',
                  'BudgetMillions','Year'])
movies.head()
```

```python
movies.info()
```

```python
movies['Genre'] = movies.Genre.astype('category')
movies['Year'] = movies.Year.astype('category')
```

```python
movies.info()
```

# Violin Plot

```python
v = sns.violinplot(data=movies, x='Genre', y='CriticsRatings')
```

# Boxplot

```python
b = sns.boxplot(data=movies, x='Genre', y='CriticsRatings')
```

# Subplot

```python
f,axies = plt.subplots(1,2,figsize=(14,4))

v = sns.violinplot(data=movies, x='Genre', y='CriticsRatings', ax= axies[0])
b = sns.boxplot(data=movies, x='Genre', y='CriticsRatings', ax= axies[1])

plt.show()
```

# DashBorad

```python
f,axies = plt.subplots(2,2,figsize=(18,10))

v = sns.violinplot(data=movies, x='Genre', y='CriticsRatings', ax= axies[0,0])
b = sns.boxplot(data=movies, x='Genre', y='CriticsRatings', ax= axies[0,1])

axies[1,0].hist(movies.AudienceRatings)
axies[1,1].hist(movies.CriticsRatings)

axies[1,0].set_xlabel('AudienceRatings')
axies[1,1].set_xlabel('CriticsRatings')

axies[1,0].set_ylabel('Probability')
axies[1,1].set_ylabel('Probability')

plt.show()
```

# FacetGrid

```python
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws = dict(edgecolor='black')
g = g.map(plt.scatter, 'CriticsRatings','AudienceRatings', **kws)
```
